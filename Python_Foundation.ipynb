# **Program for BMI!**

---




Kategori BMI.jpg

````python
# Fungsi untuk menghitung BMI
def hitung_bmi(berat_badan_kg, tinggi_badan_cm):
    # Menghitung BMI dengan rumus
    bmi = berat_badan_kg / ((tinggi_badan_cm / 100) ** 2)
    return bmi

# Input berat badan dan tinggi badan
berat_badan = float(input("Masukkan berat badan (kg): "))
tinggi_badan = float(input("Masukkan tinggi badan (cm): "))

# Memanggil fungsi hitung_bmi
bmi = hitung_bmi(berat_badan, tinggi_badan)

# Menampilkan hasil BMI
print("BMI Anda adalah:", bmi)

# Menampilkan kategori BMI
if bmi < 18.5:
    print("Kategori BMI: Kurang berat badan")
elif bmi >= 18.5 and bmi < 24.9:
    print("Kategori BMI: Normal")
elif bmi >= 25.0 and bmi < 29.9:
    print("Kategori BMI: Kelebihan berat badan")
else:
    print("Kategori BMI: Obesitas")
````

# **5. Status for BMI**


*   Normal: 18.5-24.9
*   Overweight: 25.0-29.9
*   Obesity: 30.0-34.9
*   Underweight: < 18.5

---







````python
import pandas as pd

# Membaca data awal dari file CSV
file_path = '/content/drive/MyDrive/Startupcampus/Data Assignment Kompi G - Sheet1.csv'
df = pd.read_csv(file_path)

# Menghitung BMI dan menambahkannya sebagai kolom baru
df['BMI'] = df['Berat Badan (Kg)'] / ((df['Tinggi Badan (Cm)'] / 100) ** 2)

# Fungsi untuk mengkategorikan status BMI
def categorize_bmi(bmi):
    if bmi < 18.5:
        return 'Kurang berat badan'
    elif 18.5 <= bmi < 24.9:
        return 'Normal'
    elif 25.0 <= bmi < 29.9:
        return 'Kelebihan berat badan'
    else:
        return 'Obesitas'

# Menambahkan kolom "Status BMI" dengan status yang sesuai
df['Status BMI'] = df['BMI'].apply(categorize_bmi)

# Kolom yang akan ditampilkan
df_selected = df[['No Peserta', 'BMI']]

# Menampilkan dataframe dengan kolom "Status BMI"
df_selected
````
